```python
!wget -O teste_opencv https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQAUPdTmCVVo8uzR-CzzIl9JV1YYvFPwraRuQ&s
```

Nesse exemplo utilizaremos a seguintes bibliotecas, o opencv, que √© uma biblioteca voltada pra dessenvolvimento de vis√£o computacional, midiapipe que √© uma biblioteca de machine learning para extra√ß√£o de informa√ß√µes de imagens, videos, entre outras fontes visuais, por fim o cv2_imshow do pr√≥prio que google colab, que √© uma alternativa para renderiza√ß√£o de imagens dentro da plataforma.

```python
import cv2
import  mediapipe as mp
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/teste_opencv") # Load the image directly
```

```python
# Check if the image was loaded successfully
if img is None:
    print("Error: Could not load image. Please check the file path and ensure the image exists.")
else:
    mpHands = mp.solutions.hands
    hands = mpHands.Hands()
    mpDraw = mp.solutions.drawing_utils

    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)

    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)
```

```python
cv2_imshow(img) # Display the processed image
```
